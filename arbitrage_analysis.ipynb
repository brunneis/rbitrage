# Arbitrage between Kraken and GDAX
Profit results considering a swap operation of 1 coin (LTC, ETH or BTC)

<img style="float: left;" src="https://paxforex.com/sites/default/files/images/what_is_forex_arbitrage_trading.jpg">

```python
import pandas as pd
import matplotlib.dates as mdates
import matplotlib.ticker as mticker
import matplotlib.pyplot as plt
%matplotlib inline

def fixy(ax):
    ax.xaxis.set_major_locator(mdates.MinuteLocator(interval=30))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    return ax
    
def fixx(ax):
    ax.yaxis.set_major_locator(mticker.MaxNLocator())
    return ax
```

```python
date='03122017'
```

## LTC arbitrage

```python
df_ltc = pd.read_csv("https://egr.brunneis.com/0142c2a2f13c78e9b34ad89c5166c24d7308886a4868ee72d45d5f0d7d8f4d8a/ltceur/ltceur_" + date + ".csv", parse_dates=True)
df_ltc.datetime= pd.to_datetime(df_ltc['datetime'], format='%Y-%m-%d %H:%M:%S.%f')

# Custom number of available coins (1 by default)
# available_ltc = 5
# df_ltc.profit = df_ltc.profit * available_ltc
# df_ltc.quantity = df_ltc.quantity * available_ltc
```

#### Profit vs. LTC price on GDAX and Kraken

```python
fixy(fixx(df_ltc.plot(kind='line', x='datetime', y=['profit', 'gdax', 'kraken'], figsize=(20,10), grid=True)))
```

```python
fixy(fixx(df_ltc.plot(kind='line', x='datetime', y='profit', figsize=(20,10), grid=True)))
```

#### LTC price on GDAX vs. LTC price on Kraken

```python
fixy(fixx(df_ltc.plot(kind='line', x='datetime', y=['gdax', 'kraken'], figsize=(20,10), grid=True)))
```

#### Distribution of profit for LTC

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
fixx(df_ltc.plot(kind='density', y='profit', ax=axes[0], grid=True))
fixx(df_ltc.plot(kind='box', y='profit', ax=axes[1], grid=True))
```
