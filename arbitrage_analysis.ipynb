```python
import pandas as pd
%matplotlib inline
```

```python
date='03122017'
```

# LTC arbitrage

```python
df = pd.read_csv("https://egr.brunneis.com/0142c2a2f13c78e9b34ad89c5166c24d7308886a4868ee72d45d5f0d7d8f4d8a/ltceur/ltceur_" + date + ".csv", parse_dates=True)
df.datetime= pd.to_datetime(df['datetime'], format='%Y-%m-%d %H:%M:%S.%f')
```

#### Profit vs. LTC price on GDAX and Kraken

```python
df.plot(kind='line', x='datetime', y=['profit', 'gdax', 'kraken'])
```

#### LTC price on GDAX vs. Kraken

```python
df.plot(kind='line', x='datetime', y=['gdax', 'kraken'])
```

#### Distribution of profit for LTC

```python
df.plot(kind='hist', y='profit')
```
