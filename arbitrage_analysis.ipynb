# Arbitrage between Kraken and GDAX 
Profit results considering a swap operation of 1 coin (LTC, ETH or BTC)

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.dates as mdates
import matplotlib.ticker as mticker
import matplotlib.pyplot as plt
%matplotlib inline

mpl.rcParams['xtick.labelsize'] = 16
mpl.rcParams['ytick.labelsize'] = 16

def fixy(ax):
    ax.xaxis.set_major_locator(mdates.MinuteLocator(interval=30))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
    return ax
    
def fixx(ax):
    ax.yaxis.set_major_locator(mticker.MaxNLocator())
    return ax

def get_csv_url(curr1, curr2, date):
    return f"https://egr.brunneis.com/0142c2a2f13c78e9b34ad89c5166c24d7308886a4868ee72d45d5f0d7d8f4d8a/{curr1}{curr2}/{curr1}{curr2}_{date}.csv"
```

```python
# dates=['03122017','04122017','05122017','06122017']
dates=['05122017','06122017']
```

## LTC arbitrage

```python
df_ltc = pd.concat([pd.read_csv(get_csv_url("ltc", "eur", date), parse_dates=True) for date in dates])
df_ltc.datetime= pd.to_datetime(df_ltc['datetime'], format='%Y-%m-%d %H:%M:%S.%f')
```

#### Profit vs. LTC price on GDAX and Kraken

```python
fixy(fixx(df_ltc[-2000:].plot(kind='line', x='datetime', y=['profit', 'gdax', 'kraken'], figsize=(20,10), grid=True)))
```

```python
fixy(fixx(df_ltc[-2000:].plot(kind='line', x='datetime', y='profit', figsize=(20,10), grid=True)))
```

#### LTC price on GDAX vs. LTC price on Kraken

```python
fixy(fixx(df_ltc[-2000:].plot(kind='line', x='datetime', y=['gdax', 'kraken'], figsize=(20,10), grid=True)))
```

#### Distribution of profit for LTC

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
fixx(df_ltc.plot(kind='density', y='profit', ax=axes[0], grid=True))
fixx(df_ltc.plot(kind='box', y='profit', ax=axes[1], grid=True))
```
